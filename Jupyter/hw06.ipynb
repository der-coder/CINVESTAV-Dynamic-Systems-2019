# 2nd Order Systems

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda1 = -2+sqrt(2)
lambda2 = -2-sqrt(2)
```

```python
A = np.array([[1,1,1],[lambda1+lambda2,lambda2, lambda1],[lambda1*lambda2,0,0]])
```

```python
b= np.array([[0],[1],[0.5]])
```

```python
np.linalg.inv(b)
```

```python
a = 0.5
b = -0.6035
c = 0.1035
```

```python
def myFunc(t):
    y = a +b*exp(t*lambda1) + c*exp(lambda2*t)
    return y
```

```python
myFunc(0)
```

```python
tiempo = np.linspace(0,10)
```

```python
ts=myFunc(tiempo)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plt.figure(num=1,figsize=(20,10))
plt.plot(tiempo, ts,"--*k")
plt.xlim(0, 12)
plt.legend([u'$x$'], loc=5,fontsize=20,frameon=False)
plt.xlabel(u'Tiempo', fontsize=17)
plt.tick_params(labelsize='large')
plt.ylabel(u"Soluci\\'on", fontsize=17)
plt.title(u"Time plot for x(t)",fontsize=24)
plt.tight_layout()
```
